```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

#Define sigma
def sigma(x):
    return (1 - np.exp(-100 * x)) / (1 + np.exp(-100 * x))

#Define hopfield network
def hopfield(x, t=0):
    x1, x2 = x
    dx1 = -0.5 * x1 + 0.5 * sigma(x2)
    dx2 = -0.5 * x2 + 0.5 * sigma(x1)
    return [dx1, dx2]
```

```python
#create grid and plot trajectories
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)

for x1, x2 in zip(X.ravel(), Y.ravel()):
    times = np.linspace(0, 10, 500) 
    trajectory = odeint(hopfield, [x1, x2], times)
    plt.plot(trajectory[:,0], trajectory[:,1], 'b-')

plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Phase-plane trajectories')
plt.grid(True)
plt.xlim([-1, 1])
plt.ylim([-1, 1])
plt.show()
```
